```python
#Load libraries
import pandas as pd
```

## Import datasets

```python
#cov dataset
df_cov_07 = pd.io.stata.read_stata('data/cov.dta')

df_cov_07 = df_cov_07[['pidlink', 'marstat', 'age', 'sex']]

#dl1 dataset
df_dl1_07 = pd.io.stata.read_stata('data/dl1.dta')

df_dl1_07 = df_dl1_07[['pidlink', 'dl05a', 'dl05b', 'dl06']]

#sw dataset
df_sw_14 = pd.io.stata.read_stata('data/sw.dta')

df_sw_14 = df_sw_14[['pidlink', 'sw05']]

#tk1 dataset
df_tk1_07 = pd.io.stata.read_stata('data/tk1.dta')

df_tk1_07 = df_tk1_07[['pidlink', 'hhid07', 'tk01a', 'tk01']]

# #hr1 dataset
# df_hr1_07 = pd.io.stata.read_stata('data/hr1.dta')

# df_hr1_07 = df_hr1_07[['hhid07', 'pidlink', 'hr01', 'hr1type']]

#hr1_hh dataset
df_b2_hr1_07 = pd.io.stata.read_stata('data/b2_hr1.dta')

df_b2_hr1_07 = df_b2_hr1_07[['hhid07', 'hr01', 'hrtype']]

#sc dataset
df_sc_07 = pd.io.stata.read_stata('data/sc.dta')

df_sc_07 = df_sc_07[['hhid07', 'sc010700']]

#krk dataset
df_krk_07 = pd.io.stata.read_stata('data/krk.dta')

df_krk_07 = df_krk_07.drop(['version', 'module'], axis=1)
```

df_b2_hr1_07: This is in long data format. We need to transform it into wide data in order to merge properly later on.

We also don't need multi-levels in our columns, so we'll flatten it using .to_flat_index()

```python
#transform long data into wide data
df_b2_hr1_07 = df_b2_hr1_07.pivot(columns=['hrtype'],
                     values=['hr01'],
                     index=['hhid07'])

#flatten column names
df_b2_hr1_07.columns = ["_".join(a) for a in df_b2_hr1_07.columns.to_flat_index()]

#reset index
df_b2_hr1_07.reset_index()
```

1. Create a list of dataframes to join by individual ID (pidlink)
2. Merge them all together. Call this dataframe 'df'.

```python
df_list = [df_cov_07, df_dl1_07, df_tk1_07, df_sw_14]
```

```python
from functools import reduce

#merge datasets by individual id (pidlink)
df = reduce(lambda  left,right: pd.merge(left,right,on=['pidlink'],
                                            how='left'), df_list)
```

1. Create a list of dataframes to join by household ID (hhid07)
2. Merge this with the previous dataframe, df.

```python
#merge by household id (hhid07)

df_list2 = [df, df_krk_07, df_b2_hr1_07, df_sc_07]
df = reduce(lambda  left,right: pd.merge(left,right,on=['hhid07'],
                                            how='left'), df_list2)
```

```python
#Check if any columns are missing
df.columns
```

Count total NAs in each column to see if any create problems for us by having too many NAs.

```python
# Count total NaN at each column
print(" \nCount total NaN in each column : \n\n",
      df.isnull().sum())
df.shape
```

```python
#Remove rows with NAs
df = df.dropna()

#Drop id columns
df.drop(['pidlink', 'hhid07'], axis=1, inplace=True)
```

For a first try, metrics look pretty ok, except for poor precision.

However, in the case of flagging those who are likely to fall into poverty, it is more important to correctly identify all who will fall into poverty (recall), than to have high precision (high percentage of correct classifications amongst total predicted).

**We don't really care about wrongly classifying those who won't be poor as poor, but we do care about correctly classifying those who will be poor as poor.**

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred, zero_division=True))
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(
    y_test.values.argmax(axis=1), y_pred.argmax(axis=1))
```
